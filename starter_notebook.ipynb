### Foundation Data Processing Functions

````python
import pandas as pd

# 🤖 AI Collaboration: Data Loading and Cleaning Function

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant.

    This function:
    - Loads CSV transaction data
    - Validates required columns
    - Cleans Amount values
    - Handles missing values
    - Converts data types
    - Removes invalid rows

    Args:
        file_path (str): Path to CSV file

    Returns:
        pandas.DataFrame: Cleaned transaction data
    """

    try:
        # Load CSV file
        df = pd.read_csv(file_path)

        print("✅ CSV file loaded successfully!")

        # Required columns
        required_columns = [
            "Date",
            "Amount",
            "Category",
            "Description"
        ]

        # Validate required columns
        missing_columns = [
            col for col in required_columns
            if col not in df.columns
        ]

        if missing_columns:
            raise ValueError(
                f"Missing required columns: {missing_columns}"
            )

        # Remove rows with missing values
        df = df.dropna()

        # Clean Amount column
        # Remove dollar signs and commas
        df["Amount"] = (
            df["Amount"]
            .astype(str)
            .str.replace("$", "", regex=False)
            .str.replace(",", "", regex=False)
            .str.strip()
        )

        # Convert Amount to numeric
        df["Amount"] = pd.to_numeric(
            df["Amount"],
            errors="coerce"
        )

        # Remove invalid amount rows
        df = df.dropna(subset=["Amount"])

        # Convert Date column to datetime
        df["Date"] = pd.to_datetime(
            df["Date"],
            errors="coerce"
        )

        # Remove invalid dates
        df = df.dropna(subset=["Date"])

        # Remove duplicate transactions
        df = df.drop_duplicates()

        # Reset index
        df = df.reset_index(drop=True)

        # Business summary
        print("✅ Data cleaned successfully!")
        print(f"📊 Total transactions: {len(df)}")
        print(f"💰 Total spending: ${df['Amount'].sum():.2f}")

        return df

    except FileNotFoundError:
        print("❌ File not found. Please check the file path.")

    except pd.errors.EmptyDataError:
        print("❌ The CSV file is empty.")

    except ValueError as ve:
        print(f"❌ Validation Error: {ve}")

    except Exception as e:
        print(f"❌ Unexpected Error: {e}")
````

````python
df_sample.to_csv("sample_transactions (1).csv", index=False)
clean_data = load_and_clean_transaction_data(
    "sample_transactions (1).csv"
)

print(clean_data.head())
````

````python
# 🤖 AI Collaboration: Spending Analysis Function

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights.

    Args:
        df (pandas.DataFrame):
            Cleaned transaction DataFrame

    Returns:
        dict:
            Analysis results and business insights
    """

    try:
        # Calculate total spending
        total_spending = df["Amount"].sum()

        # Spending by category
        category_totals = (
            df.groupby("Category")["Amount"]
            .sum()
            .sort_values(ascending=False)
        )

        # Percentage by category
        category_percentages = (
            (category_totals / total_spending) * 100
        ).round(2)

        # Average transaction
        average_transaction = df["Amount"].mean()

        # Highest spending category
        top_category = category_totals.idxmax()

        # Highest spending amount
        top_category_amount = category_totals.max()

        # Generate business insights
        insights = []

        insights.append(
            f"Total spending is ${total_spending:.2f}."
        )

        insights.append(
            f"The highest spending category is "
            f"{top_category} "
            f"with ${top_category_amount:.2f} spent."
        )

        insights.append(
            f"The average transaction amount is "
            f"${average_transaction:.2f}."
        )

        # Savings recommendation
        if category_percentages.iloc[0] > 40:
            insights.append(
                f"{top_category} represents more than 40% "
                f"of total spending. Consider reviewing "
                f"expenses in this category."
            )

        # Build results dictionary
        analysis_results = {
            "total_spending": total_spending,
            "average_transaction": average_transaction,
            "category_totals": category_totals,
            "category_percentages": category_percentages,
            "top_category": top_category,
            "business_insights": insights
        }

        # Display formatted summary
        print("✅ Spending Analysis Completed!\n")

        print(f"💰 Total Spending: ${total_spending:.2f}")
        print(
            f"📊 Average Transaction: "
            f"${average_transaction:.2f}"
        )
        print(f"🏆 Top Category: {top_category}")

        print("\n📂 Spending by Category:")
        print(category_totals)

        print("\n📈 Category Percentages:")
        print(category_percentages)

        print("\n💡 Business Insights:")
        for insight in insights:
            print(f"- {insight}")

        return analysis_results

    except Exception as e:
        print(f"❌ Error analyzing spending patterns: {e}")
````

````python
analysis = analyze_spending_patterns(clean_data)
````

````python
# 🤖 AI Collaboration: Business Insights Generator

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations
    based on spending analysis.

    Args:
        analysis_data (dict):
            Dictionary containing spending analysis results

    Returns:
        str:
            Formatted recommendation report
    """

    try:
        # Extract analysis data
        total_spending = analysis_data["total_spending"]
        average_transaction = analysis_data["average_transaction"]
        top_category = analysis_data["top_category"]
        category_percentages = analysis_data["category_percentages"]

        # Create recommendation list
        recommendations = []

        recommendations.append(
            "SMART FINANCE ASSISTANT REPORT"
        )

        recommendations.append("=" * 40)

        recommendations.append(
            f"Total Spending: ${total_spending:.2f}"
        )

        recommendations.append(
            f"Average Transaction Amount: "
            f"${average_transaction:.2f}"
        )

        recommendations.append(
            f"Top Spending Category: {top_category}"
        )

        recommendations.append("\nFINANCIAL INSIGHTS")

        # Analyze category percentages
        for category, percentage in category_percentages.items():

            recommendations.append(
                f"- {category}: {percentage:.2f}% "
                f"of total spending"
            )

            # Business recommendations
            if percentage > 40:
                recommendations.append(
                    f"  Recommendation: Spending in "
                    f"{category} is very high. "
                    f"Consider setting a monthly budget "
                    f"limit for this category."
                )

            elif percentage > 25:
                recommendations.append(
                    f"  Recommendation: Monitor spending "
                    f"in {category} to maintain a balanced "
                    f"financial plan."
                )

        # Savings suggestions
        recommendations.append("\nSAVINGS OPPORTUNITIES")

        if average_transaction > 100:
            recommendations.append(
                "- Large average transaction amounts "
                "may indicate opportunities to reduce "
                "non-essential purchases."
            )

        recommendations.append(
            "- Track recurring expenses monthly."
        )

        recommendations.append(
            "- Consider setting category spending goals."
        )

        recommendations.append(
            "- Review subscriptions and unnecessary expenses."
        )

        recommendations.append(
            "- Build an emergency savings fund "
            "for financial stability."
        )

        # Final summary
        recommendations.append("\nFINAL SUMMARY")

        recommendations.append(
            "Your spending analysis has been completed "
            "successfully. Regular monitoring of spending "
            "patterns can improve budgeting decisions "
            "and long-term financial health."
        )

        # Convert list to formatted string
        final_report = "\n".join(recommendations)

        return final_report

    except Exception as e:
        return f"❌ Error generating recommendations: {e}"
````

## Custom Financial Tools

````python
# 🤖 AI Collaboration: Custom Tool Development

from hands_on_ai import agent
import math

def create_savings_calculator_tool(
    current_savings,
    monthly_contribution,
    target_amount
):
    """
    Calculate time required to reach a savings goal.

    Args:
        current_savings (float):
            Current amount saved

        monthly_contribution (float):
            Monthly savings contribution

        target_amount (float):
            Savings goal target

    Returns:
        str:
            Formatted savings analysis
    """

    try:
        # Input validation
        if current_savings < 0:
            return "❌ Current savings cannot be negative."

        if monthly_contribution <= 0:
            return (
                "❌ Monthly contribution must be greater than zero."
            )

        if target_amount <= 0:
            return "❌ Target amount must be greater than zero."

        if current_savings >= target_amount:
            return (
                "✅ You have already reached your savings goal!"
            )

        # Calculate remaining amount
        remaining_amount = (
            target_amount - current_savings
        )

        # Calculate months required
        months_required = math.ceil(
            remaining_amount / monthly_contribution
        )

        # Convert months to years/months
        years = months_required // 12
        months = months_required % 12

        # Build formatted report
        report = []

        report.append(
            "💰 SAVINGS GOAL CALCULATOR"
        )

        report.append("=" * 35)

        report.append(
            f"Current Savings: "
            f"${current_savings:,.2f}"
        )

        report.append(
            f"Monthly Contribution: "
            f"${monthly_contribution:,.2f}"
        )

        report.append(
            f"Target Savings Goal: "
            f"${target_amount:,.2f}"
        )

        report.append(
            f"Remaining Amount: "
            f"${remaining_amount:,.2f}"
        )

        report.append(
            f"\nEstimated Time to Reach Goal:"
        )

        report.append(
            f"{months_required} months"
        )

        # Add year/month breakdown
        if years > 0:
            report.append(
                f"Approximately {years} year(s) "
                f"and {months} month(s)"
            )

        # Financial advice
        report.append("\n📈 FINANCIAL ADVICE")

        if months_required > 60:
            report.append(
                "- Consider increasing monthly contributions "
                "to reach your goal faster."
            )

        elif months_required < 12:
            report.append(
                "- Excellent progress! Your goal is achievable "
                "within one year."
            )

        else:
            report.append(
                "- Stay consistent with your monthly savings habit."
            )

        report.append(
            "- Review spending patterns regularly "
            "to identify extra savings opportunities."
        )

        # Return formatted string
        return "\n".join(report)

    except Exception as e:
        return f"❌ Error calculating savings goal: {e}"


# Register tool with agent system
agent.register_tool(
    "savings_calculator",
    "A calculator that determines the time required to reach a savings goal. Requires `current_savings` (float), `monthly_contribution` (float), and `target_amount` (float). Returns a formatted string with estimated time and financial advice.",
    create_savings_calculator_tool
)
````

````python
# 🤖 AI Collaboration: Professional UI Design

import gradio as gr
import pandas as pd


def create_finance_assistant_ui():

    # CSV Analysis Function
    def process_finance_file(file):

        try:
            # Load and clean data
            clean_data = load_and_clean_transaction_data(
                file.name
            )

            # Analyze spending
            analysis = analyze_spending_patterns(
                clean_data
            )

            # Generate recommendations
            recommendations = (
                generate_financial_recommendations(
                    analysis
                )
            )

            return recommendations

        except Exception as e:
            return f"❌ Error processing file: {e}"

    # Savings Calculator Wrapper
    def calculate_savings(
        current_savings,
        monthly_contribution,
        target_amount
    ):

        return create_savings_calculator_tool(
            current_savings,
            monthly_contribution,
            target_amount
        )

    # Chatbot Function
    def finance_chat(user_message):

        try:
            finance_personality = (
                create_finance_chat_personality()
            )

            response = chat.say(
                user_message,
                system_prompt=finance_personality
            )

            return response

        except Exception:

            # Fallback response
            return (
                "I recommend reviewing your monthly "
                "spending habits, reducing unnecessary "
                "expenses, and setting clear savings goals."
            )

    # Main Gradio Interface
    with gr.Blocks(
        title="Smart Finance Assistant"
    ) as demo:

        gr.Markdown(
            """
            # 💰 Smart Finance Assistant

            Analyze spending habits, generate financial
            insights, chat with an AI finance assistant,
            and calculate savings goals.
            """
        )

        # TAB 1 — Spending Analysis
        with gr.Tab("📊 Spending Analysis"):

            file_input = gr.File(
                label="Upload Transaction CSV File"
            )

            analysis_output = gr.Textbox(
                label="Financial Analysis Report",
                lines=25
            )

            analyze_button = gr.Button(
                "Analyze Spending"
            )

            analyze_button.click(
                fn=process_finance_file,
                inputs=file_input,
                outputs=analysis_output
            )

        # TAB 2 — Finance Chatbot
        with gr.Tab("🤖 Finance Chatbot"):

            chatbot_input = gr.Textbox(
                label="Ask a Financial Question",
                placeholder=(
                    "Example: How can I reduce "
                    "my food expenses?"
                )
            )

            chatbot_output = gr.Textbox(
                label="AI Financial Advice",
                lines=10
            )

            chat_button = gr.Button(
                "Get Advice"
            )

            chat_button.click(
                fn=finance_chat,
                inputs=chatbot_input,
                outputs=chatbot_output
            )

        # TAB 3 — Savings Calculator
        with gr.Tab("💰 Savings Calculator"):

            current_savings = gr.Number(
                label="Current Savings",
                value=0
            )

            monthly_contribution = gr.Number(
                label="Monthly Contribution",
                value=100
            )

            target_amount = gr.Number(
                label="Target Savings Goal",
                value=1000
            )

            savings_output = gr.Textbox(
                label="Savings Goal Report",
                lines=15
            )

            calculate_button = gr.Button(
                "Calculate Savings Goal"
            )

            calculate_button.click(
                fn=calculate_savings,
                inputs=[
                    current_savings,
                    monthly_contribution,
                    target_amount
                ],
                outputs=savings_output
            )

        # Footer
        gr.Markdown(
            """
            ---
            Smart Finance Assistant
            AI-Powered Personal Finance Analysis Tool
            """
        )

    return demo


# Launch application
demo = create_finance_assistant_ui()
demo.launch()
````
